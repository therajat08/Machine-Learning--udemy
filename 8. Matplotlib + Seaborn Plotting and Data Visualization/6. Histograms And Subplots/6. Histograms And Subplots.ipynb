```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nut_butter_prices = {"almond butter":10,
                    "peanut butter":9,
                    "cashew butter":34}
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
```

```python
#make some data
#randn gives  random distribution numbers
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# creating subplot option 1
#creating multiple plots at the same time
fig, ((ax1, ax2),(ax3,ax4))= plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
#plot to each different different axis

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(2000));
```
