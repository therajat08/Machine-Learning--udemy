## making figures with numpy arrays

We want
* line plot
* subplot
* scatter plot
* histogram
* bar plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create data
x = np.linspace(0,10,100)
#returned evenly spaced numbers b/w 0 and 10
```

```python
x
```

```python
# plotting data
fig, ax = plt.subplots()
ax.plot(x,x**2)
#note the default plot is line type unlike others llike bar, scatter
```

```python
#scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# similar
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# make from a dictionary
nut_butter_prices = {"almond butter":10,
                    "peanut butter":9,
                    "cashew butter":34}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="fav butters",
      ylabel="prices");
```

```python
#histogram
x = np.random.randn(1000)
#randn() returns samples from normal distribution :{symmetric figures, google it}
fig,ax=plt.subplots()
ax.hist(x);
```
