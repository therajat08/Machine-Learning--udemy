```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#subplots option 2
#earlier was tuple way; this is list-index way
x = np.random.randn(1000)
nut_butter_prices = {"almond butter":10,
                    "peanut butter":9,
                    "cashew butter":34}

fig , ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))

ax[0,0].plot(x,x/2); #[0,0] means top left figure in coordinate system
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(2000));
```
